```python
ZonaProp_data = ZonaProp_data[ZonaProp_data['precio'] != 'N/A']
ZonaProp_data.head()
```

```python
ZonaProp_data = ZonaProp_data.drop_duplicates(subset=['precio', 'localizacion'])
ZonaProp_data[['precio', 'localizacion']].sort_values('precio', ascending=False).head(100)
```

```python
ZonaProp_data.hist(column='precio', bins=100)
```

```python
ZonaProp_data = ZonaProp_data[ZonaProp_data['precio'] < 350000]
```

```python
ZonaProp_data.hist('precio', bins=75)
```

```python
ZonaProp_data = ZonaProp_data.drop('id', axis=1)
ZonaProp_data.describe()
```

```python
import scipy.stats as stats
```

```python
data = ZonaProp_data['precio'].sample(n=30)
ks_statistic, p_value = stats.kstest(data, 'norm', args=(125256, 78661))

# Print the results
print(f"KS Statistic: {ks_statistic}")
print(f"P-value: {p_value}")
```

```python
# Interpretation of results
if p_value > 0.05:
    print("The sample data follows a normal distribution (fail to reject H0).")
else:
    print("The sample data does not follow a normal distribution (reject H0).")
```
